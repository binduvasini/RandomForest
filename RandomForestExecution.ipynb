```python
import pandas as pd
from decision_tree import decision_tree
from random_forest import random_forest
import random
from sklearn.metrics import accuracy_score
import numpy as np
import math
from sklearn import preprocessing
```

```python
tree_list = []
def fit(X, Y, trees=10):
    feat = X.columns.size+1
    features = list(range(1,feat))
    for t in range(trees):
        sample_x = X.sample(n=X.size,replace=True)
        sample_x_values = sample_x.values.tolist()
        sample_y_values = [Y[idxl] for idxl in sample_x.index.values]
        feature_list = random.sample(features, math.floor(math.sqrt(feat)))
        r = random_forest()
        the_tree = r.fit(sample_x_values, sample_y_values, feature_list)
        tree_list.append(r)
```

```python
def predict(X):
    from collections import defaultdict
    hypotheses = [tr.predict(X) for tr in tree_list]  #for each row in X, we get hypothesis which will be appended to a list
    returnhyp = []
    for i in range(len(X)):
        counts = defaultdict(int)
        for h in hypotheses: #we get the list of hypotheses list
            counts[h[i]] += 1
        major = max(counts, key = counts.get)
        returnhyp.append(major)
    return returnhyp
```
